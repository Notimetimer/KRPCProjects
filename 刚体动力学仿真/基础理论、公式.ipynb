## 飞行器动力学和运动学方程

### 重要事项：
- 先算旋转，再算位移。
- 对于物理引擎，飞行器的运动学计算/导航计算一律使用惯性系，在体轴系计算运动会导致积分误差迅速积累，且辛普森积分或者龙格库塔积分也无法改善（梯形积分和辛普森积分甚至会加剧积分误差恶化）
- 约定在前头，为了保证公式简洁，写的一律是矢量/矩阵方程，矢量一律为列向量。
- 公式来自豆包


#### 体轴系动力学方程
1. 体轴系下旋转动力学方程
   $$\mathbf{J_b}\dot{\boldsymbol{\omega_b}}+\boldsymbol{\omega_b}\times(\mathbf{J_b}\boldsymbol{\omega_b})=\boldsymbol{M_b}$$
   其中 $\boldsymbol{\omega_b}$ 是机体相对惯性系的角速度在体轴系的投影（或者说惯性系相对机体系的角速度的反向），$\mathbf{J_b}$ 是惯性张量矩阵，好在在体轴系下面是常量。$\boldsymbol{M_b}$是体轴系下的力矩，角动量对时间求导总是包含 转速的导数 * 惯性张量 和 惯性张量的导数 * 转速，后者为陀螺力矩。

2. 体轴系下线运动动力学方程
   $$\dot{\boldsymbol{V}}_{b}=\boldsymbol{V}_{b}\times\boldsymbol{\omega}_{b}+\frac{1}{m}\boldsymbol{F}_{b}$$
   力项是合外力，除了 F/m， 体轴系的受力还得考虑牵连的向心加速度，动力学方程写在质心，所以消去科氏加速度和欧拉加速度。

体轴系计算出来的直接用在控制率上面，不要在体轴系上写运动学方程

#### 惯性系动力学方程
1. 惯性系下旋转动力学方程
   $$\mathbf{J_i}\dot{\boldsymbol{\omega_i}}+\boldsymbol{\omega_i}\times(\mathbf{J_i}\boldsymbol{\omega_i})=\boldsymbol{M_i}$$
   这里的 $\mathbf{J_i}$ 可不是常数，而是时变的量，和坐标变换矩阵有关： $\mathbf{J_i} = \boldsymbol{R_{bi}} \cdot \mathbf{J_b} \cdot \boldsymbol{R_{bi}}^T$

2. 惯性系下的线运动动力学方程
   $$\dot{\boldsymbol{V}}_{i}=\frac{1}{m}\boldsymbol{F}_{i}$$
   没有牵连运动就是这么简洁

#### 惯性系运动学方程
1. 旋转部分：
   路径1、先积分角速度矢量得到旋转四元数（含旋转轴和旋转角度），然后用罗德里格斯公式更新姿态
   路径2、直接使用体轴系下的p,q,r积分四元数，但是会有误差（步长不能太大），而且需要频繁归一化四元数
   $$\begin{Bmatrix}\dot{q}_0\\\\\dot{q}_1\\\\\dot{q}_2\\\\\dot{q}_3\end{Bmatrix}=\frac{1}{2}\begin{Bmatrix}0&-p&-q&-r\\\\p&0&r&-q\\\\q&-r&0&p\\\\r&q&-p&0\end{Bmatrix}\begin{Bmatrix}q_0\\\\q_1\\\\q_2\\\\q_3\end{Bmatrix}$$

2. 质心运动部分：
   $$\dot{\boldsymbol{R_i}} = \boldsymbol{V_i}$$
    真的要写吗？几乎就是高中物理的矢量化，再加上数值积分方法
   

物理引擎建议计算路径：
1. 所有力和力矩都搬到惯性系来（直接做向量的坐标变换）
2. 在惯性系下先积分角速度得到旋转轴和角度，再利用罗德里格斯旋转公式解析更新姿态（无线性化误差）
3. 在惯性系下积分质心运动，不容易累积积分误差

## 多连杆动力学与运动学方程

计算路径：
- 动力学：由末梢到根部传递
- 运动学：由根部到末梢

假设杆子编号从0,1,2,...增大，对中间的杆n:
1. 质心加速度分解/合成公式（写在惯性系）：
    $$\mathbf{a}_{c,n}=\underbrace{\mathbf{a}_{n}}_{\text{牵连平动}}+\underbrace{\dot{\boldsymbol{\omega}}_{n}\times\mathbf{r}_{c,n}}_{\text{欧拉}}+\underbrace{\boldsymbol{\omega}_{n}\times(\boldsymbol{\omega}_{n}\times\mathbf{r}_{c,n})}_{\text{离心}}+\underbrace{2\boldsymbol{\omega}_{n}\times\dot{d}_{n}\mathbf{u}_{n}}_{\text{科里奧利}}+\underbrace{\ddot{d}_{n}\mathbf{u}_{n}}_{\text{伸缩}}$$

2. 动力学（由末梢向中心传，把力写在惯性系上）：
    1. 力(并非完整合力)
    $$\mathbf{F}_n=m_n\mathbf{a}_{c,n}$$
    2. 力矩(并非完整合力矩)
    $$\mathbf{M}_n=\mathbf{J}_n\dot{\boldsymbol{\omega}}_n+\boldsymbol{\omega}_n\times(\mathbf{J}_n\boldsymbol{\omega}_n)$$
    3. 关节力/力矩递推(写在杆子末梢上)
    $$\mathbf{F}_{n}=\mathbf{F}_{n+1}+\mathbf{F}_{ext,n}$$
    $$\mathbf{M}_{n}=\mathbf{M}_{n+1}+\mathbf{M}_{ext,n}+\mathbf{r}_n\times\mathbf{F}_n$$
