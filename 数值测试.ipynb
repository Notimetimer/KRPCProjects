```python
import krpc  
import time 
import numpy as np
import math  
from math import *
from coord_rotations import *


# 左手系速度（0E, N, 90E）
velocity0 = np.array([
    20  , 
    0  , 
    0 ,
    ])
longitude = 270 # 经度
latitude = 90 # 纬度 注意在两极点经度没有意义，会导致很诡异的结果


# z轴取反，变为右手系（0E, N, -90E）
velocity0[2] = - velocity0[2]
# 两极附近强制经度归零
if abs(latitude) > 90 - 1e-1:
    latitude = 0
# 坐标旋转变换，函数里面有psi=-heading，但这里经度正是y轴对应的旋转角
velocity0 = passive_rotation(velocity0, heading=-longitude*pi/180, theta=latitude*pi/180)
# 上、（地表）北、西速度再转回北天东
velocity = np.zeros(3)
velocity[0] = velocity0[1]
velocity[1] = velocity0[0]
velocity[2] = -velocity0[2]

print("北天东速度", np.round(velocity,2))
```
